```python
import icalendar
```

```python
# use icalendar to parse the academic-calendar.ics file and convert the events into a dictionary
with open('academic-calendar.ics', 'rb') as f:
    gcal = icalendar.Calendar.from_ical(f.read())
```

```python
# print all the events in the calendar
for component in gcal.walk():
    if component.name == "VEVENT":
        print(component.get('summary'))
        print(component.get('dtstart').dt)
        print(component.get('dtend').dt)
```
